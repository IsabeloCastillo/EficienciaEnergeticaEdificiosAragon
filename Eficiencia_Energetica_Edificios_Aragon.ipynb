#### Faltan 316 lineas con fechas de construccion, vamos hacer un exploracion para identificar si tienen correlacion , es decir si todas esos valores faltantes son de una misma localidad, provincia o tipo en cuestion para que no nos haga un sesgo.

#### Las columnas Emision_CO2 y ConsumoKWh/m2/Anio parecen ser numéricas, pero están marcadas como tipo objeto. Probablemente porque los números se registran como cadenas de texto, posiblemente con comas para los decimales o unidades incluidas. Tendremos que convertir estas a tipo flotante y asegurarnos de que el punto decimal esté correctamente posicionado.

```python
# Realiza tus operaciones
df_sinanio['Emision_CO2'] = df_sinanio['Emision_CO2'].str.replace(',', '.').str.extract('(\d+\.?\d*)').astype(float)
df_sinanio['ConsumoKWh/m2/Anio'] = df_sinanio['ConsumoKWh/m2/Anio'].str.replace(',', '.').str.extract('(\d+\.?\d*)').astype(float)

# Selecciona las columnas de interés y muestra las primeras filas
df_sinanio[['Emision_CO2', 'ConsumoKWh/m2/Anio']].head()
```

Emision_CO2, Anio_construccion y Superficie_m2, tienen valore atipicos en su valor **max**, procedemos a su correción

```python
import pandas as pd

# Suponiendo que ya has cargado tu DataFrame en 'df'

# Establece un umbral para cada columna donde identificaste valores atípicos
umbral_emision_CO2 = 300
umbral_anio_construccion = 2025
umbral_superficie = 5000

# Elimina las filas que superen el umbral
df_sinanio = df_sinanio[
    (df_sinanio['Emision_CO2'] < umbral_emision_CO2) &
    (df_sinanio['Anio_construccion'] < umbral_anio_construccion) &
    (df_sinanio['Superficie_m2'] < umbral_superficie)
]

# Verifica los cambios con describe()
df_sinanio.describe()
```

#### La columna coordenadas también está listada como objeto. Si contiene datos separados por comas, como latitud y longitud, querremos dividirla en dos columnas numéricas separadas.

```python
import pandas as pd
import numpy as np

# Asegurarse de que solo se incluyen filas con el formato correcto en 'Coordenadas_gps'
df_sinanio_filtrado = df_sinanio[df_sinanio['Coordenadas_gps'].str.count(',') == 1]

# Dividir la columna 'Coordenadas_gps' en dos nuevas columnas 'Latitud' y 'Longitud'
df_sinanio_filtrado[['Latitud', 'Longitud']] = df_sinanio_filtrado['Coordenadas_gps'].str.split(',', expand=True)

# Reemplazar cadenas vacías con NaN y luego convertir a flotante
df_sinanio_filtrado['Latitud'] = pd.to_numeric(df_sinanio_filtrado['Latitud'].replace('', np.nan), errors='coerce')
df_sinanio_filtrado['Longitud'] = pd.to_numeric(df_sinanio_filtrado['Longitud'].replace('', np.nan), errors='coerce')

# Verificar los cambios realizados
print(df_sinanio_filtrado[['Latitud', 'Longitud']].head())
```

Además, podríamos calcular el coeficiente de correlación para entender la fuerza y la dirección de la relación entre las dos variables.

#### Este código dará la correlación de Pearson, que mide la relación lineal entre ConsumoKWh/m2/Anio y Emision_CO2. Un valor cercano a 1 o -1 indica una fuerte relación lineal, mientras que un valor cercano a 0 indica una relación lineal débil o inexistente.

```python
# Calculamos la correlación de Pearson
correlacion = df_sinanio[['ConsumoKWh/m2/Anio', 'Emision_CO2']].corr()

print(correlacion)
```

La matriz de correlación muestra que la correlación entre ConsumoKWh/m2/Anio y Emision_CO2 es extremadamente baja, prácticamente cero, lo que indica que no hay una relación lineal aparente entre estas dos variables en el conjunto de datos filtrado. Esto podría deberse a una variedad de factores, incluyendo la naturaleza de los datos, posibles valores atípicos restantes, o simplemente que no existe una relación lineal directa entre estas variables.

#### Filtraremos la variable ConsumoKWh/m2/Anio de la misma manera que hicimos con Emision_CO2. Esto podría aclarar si hay alguna relación subyacente.

```python
# Calculamos los cuartiles y el rango intercuartílico (IQR)
Q1 = df_sinanio['ConsumoKWh/m2/Anio'].quantile(0.25)
Q3 = df_sinanio['ConsumoKWh/m2/Anio'].quantile(0.75)
IQR = Q3 - Q1

# Definimos los límites para considerar un dato como atípico
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtramos los datos para remover los valores atípicos
df_filtrado_sin_atipicos = df_sinanio[(df_sinanio['ConsumoKWh/m2/Anio'] >= limite_inferior) & 
                                        (df_sinanio['ConsumoKWh/m2/Anio'] <= limite_superior)]

# Mostramos la descripción de la columna filtrada para verificar
df_filtrado_sin_atipicos['ConsumoKWh/m2/Anio'].describe()
```

Con los nuevos datos filtrados de la variable ConsumoKWh/m2/Anio, hemos eliminado los valores atípicos y ahora tenemos un conjunto de datos más coherente con una desviación estándar reducida y un rango más estrecho de valores.

```python
# Calculamos la correlación de Pearson en el conjunto de datos filtrado
correlacion_filtrada = df_filtrado_sin_atipicos[['ConsumoKWh/m2/Anio', 'Emision_CO2']].corr()

print(correlacion_filtrada)
```

Esta vez, la correlación entre ConsumoKWh/m2/Anio y Emision_CO2 es muy alta, lo que indica una fuerte relación lineal positiva entre estas dos variables en el conjunto de datos filtrado. Esto significa que, en general, a medida que el consumo de energía por metro cuadrado por año aumenta, las emisiones de CO2 también tienden a aumentar, y viceversa.

### Ahora que hemos establecido una fuerte correlación lineal entre las dos variables, sería valioso visualizar esta relación con un gráfico de dispersión. También podríamos ajustar una línea de tendencia para destacar la relación lineal.

```python
import matplotlib.pyplot as plt
import numpy as np

# Gráfico de dispersión para ConsumoKWh/m2/Anio vs Emision_CO2
plt.figure(figsize=(10, 6))
plt.scatter(df_filtrado_sin_atipicos['ConsumoKWh/m2/Anio'], df_filtrado_sin_atipicos['Emision_CO2'], alpha=0.5)

# Ajustamos una línea de tendencia lineal
m, b = np.polyfit(df_filtrado_sin_atipicos['ConsumoKWh/m2/Anio'], df_filtrado_sin_atipicos['Emision_CO2'], 1)
plt.plot(df_filtrado_sin_atipicos['ConsumoKWh/m2/Anio'], m*df_filtrado_sin_atipicos['ConsumoKWh/m2/Anio'] + b, color='red')

plt.title('Relación entre Consumo de Energía y Emisiones de CO2')
plt.xlabel('ConsumoKWh/m2/Anio')
plt.ylabel('Emisiones de CO2')
plt.grid(True)
plt.show()
```

El gráfico de dispersión actualizado muestra claramente la fuerte relación lineal entre el consumo de energía por metro cuadrado al año y las emisiones de CO2. La línea roja representa la línea de tendencia que confirma visualmente la correlación que calculamos anteriormente.

La nube de puntos y la línea de tendencia indican que, en general, un mayor consumo de energía está asociado con mayores emisiones de CO2, lo cual tiene sentido desde un punto de vista práctico.

# CONCLUSIONES

## Relación entre Consumo de Energía y Emisiones de CO2:

Hay una correlación significativa entre el consumo de energía y las emisiones de CO2, lo que indica que las estrategias para reducir el consumo de energía podrían ser efectivas en la disminución de las emisiones.

## Impacto de la Antigüedad del Edificio:

La antigüedad del edificio, representada por el año de construcción, no muestra una correlación directa con el consumo de energía ni con las emisiones de CO2. Esto sugiere que otros factores, como las renovaciones, el mantenimiento y la implementación de tecnologías eficientes, podrían ser más influyentes en la eficiencia energética de un edificio que el año de construcción en sí.

## Variabilidad en las Características del Edificio:

Existe una amplia variabilidad en las emisiones de CO2 y en el consumo de energía entre los edificios, lo que puede reflejar una diversidad de usos, eficiencias de sistemas de energía y prácticas de construcción.

## Identificación de Valores Atípicos:

Los valores atípicos presentes en las variables Emision_CO2 y ConsumoKWh/m2/Anio inicialmente distorsionaron la comprensión de la distribución de estos datos. Su filtrado nos permitió obtener una perspectiva más clara y precisa de la mayoría de los edificios.

## Distribución de la Superficie del Edificio:

No se observó una relación fuerte entre la superficie del edificio y las emisiones o el consumo de energía, lo que indica que el tamaño del edificio no es necesariamente un predictor de mayor consumo o emisiones.

## Patrones Temporales:

## Distribuciones Unimodales:

Las distribuciones unimodales de las emisiones de CO2 y del consumo de energía sugieren que la mayoría de los edificios tienen características de eficiencia energética similares, con un menor número de edificios que se desvían significativamente de la norma.

Estas conclusiones forman un conjunto integral de observaciones que pueden servir como base para recomendaciones específicas o para el desarrollo de políticas destinadas a mejorar la eficiencia energética y reducir las emisiones de CO2. Por ejemplo, podrían justificar la inversión en tecnologías de ahorro de energía o en programas de renovación de edificios más antiguos. Además, estos insights pueden guiar la investigación futura para comprender mejor las variables que influyen en la eficiencia energética y las emisiones en diferentes tipos de edificios.
